```python
%%timeit
client = list(df2['session_id'])
topic = list(df2['text_label'])
def dub_n(arr):
    i = 0
    tmp = arr[i]
    c = float('nan')
    save = []
    for item in arr:
#         print(item, item, arr[0])
#         print(c)
        if tmp == item and c == client[i]:
            save.append(c)
        c = client[i]
        tmp = item
        i += 1
    return save

dublicate = dub_n(topic)
# len(dublicate)
```

```python
def iterat_popul(goalist):
    c = None
    fix = {}
#     print(goalist[0])
    for item in goalist:
        if item == c:
#             print('f')
            fix[item] = fix.get(item, 0) + 1
#         print('x')
        c = item
    return fix
```

```python
fin_dic = iterat_popul(list(df['text_label']))
list(dict(sorted(fin_dic.items(), key=lambda item: item[1], reverse=True)))[:10]
```

```python
client = list(df2['session_id'])
def popul_in_ses(arr):
    fix = {}
    c = None
    i = 0
    for item in arr:
        if c == client[i]:
            fix[item] = fix.get(item, 0) + 1
        c = client[i]
        i += 1
    return fix
```

```python
fin_dic2 = popul_in_ses(list(df2['text_label']))
list(dict(sorted(fin_dic2.items(), key=lambda item: item[1], reverse=True)))[:10]
```
